```python
import urllib.request
import urllib.parse
import os

url = 'http://tieba.baidu.com/f?ie=utf-8'

#输入吧名 输入起始页码 结束页码 然后在当前文件夹中创建一个一吧名为名字的文件夹
#每一页是html内容   文件名是  吧名_page.html
ba_name = input('请输入要爬取得吧名：')
start_page = int(input('请输入要访问起始页：'))
end_page = int(input('请输入要访问结束页：'))
number = 50
if not os.path.exists(ba_name):
    os.mkdir(ba_name)
headers = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.86 Safari/537.36',
}
#循环 依次爬去每一页
for page in range(start_page, end_page + 1):
    #page就是当前页
    #拼接url的过程
    data = {
        'kw': ba_name,
        'pn': (page - 1) * 50
    }
    data = urllib.parse.urlencode(data)
    url_t = url + data

    request = urllib.request.Request(url_t, headers = headers)
    print("第%s页开始下载。。。。" % page)
    response = urllib.request.urlopen(request)
    #创建文件夹和文件
    filename = ba_name + '_' + str(page) + '.html'
    #拼接文件路径
    filepath = ba_name + '/' + filename

    with open(filepath, 'wb') as fp:
        fp.write(response.read())
```

```python
#Handler处理器、自定义Opener
#urlopen()  给一个url，发送请求， 获取响应

#Request()   定制请求头， 创建请求对象
#高级功能： 使用代理， cookie 
#基本用法
#handler
import urllib.request
import urllib.parse

url = 'http://www.baidu.com/'

headers = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.86 Safari/537.36',
}

#第一步创建一个handler
handler = urllib.request.HTTPHandler()

#通过handler创建一个opener
#opener就是一个对象，发送请求直接使用opener里的方法即可
opener = urllib.request.build_opener(handler)

#构建请求对象

request = urllib.request.Request(url, headers = headers)
response = opener.open(request)

print(response.read().decode())
```

```python
#图片下载
import urllib.request 
import urllib.parse
import re
import os
import time
def handle_request(url, page):
    url = url + str(page) + '/'
    #print(url)
    headers = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.86 Safari/537.36',
    }
    request = urllib.request.Request(url = url, headers = headers)
    return request

def download_image(content):
    pattern = re.compile(r'<div class="thumb">.*?<img src="(.*?)" .*?>.*?</div>', re.S)
    lt = pattern.findall(content)
    #print(len(lt))
    #遍历lt依次下载
    for url_image in lt:
        #先处理image
        url_image = 'https:' + url_image
        #发送请求下载图片
        #创建文件夹
        dirname = 'qiutu'
        if  not os.path.exists(dirname):
            os.mkdir(dirname)
        urllib.request.urlretrieve(url_image)
        filename = url_image.split('/')[-1]
        filepath = dirname + '/' + filename
        print('%s图片正在下载。。。' % filename)
        urllib.request.urlretrieve(url_image, filepath)
        print('%s图片结束下载。。。' % filename)

def main():
    url = 'https://www.qiushibaike.com/pic/page/'
    start_page = int(input('请输入起始页码：'))
    end_page = int(input('请输入结束页码：'))
    for page in range(start_page, end_page + 1):
        #生成请求对象
        print("第%d页开始下载。。。。" % page)
        request = handle_request(url, page)
        content = urllib.request.urlopen(request).read().decode()
        #解析内容 提取所有的图片链接
        time.sleep(1)
        
        download_image(content)
a = main()





```

```python
#爬取名人名言  爬取指定页面的需求
import urllib.request
import urllib.parse
import re
import os

def handle_request(url, page = None):
    if page != None:
        #拼接处指定的url
        url = url + str(page) + '.html'
    headers = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.86 Safari/537.36',
    }
    request = urllib.request.Request(url = url, headers=headers)
    return request

def get_text(a_href):
    request = handle_request(a_href)
    #发送请求，获得响应
    content = urllib.request.urlopen(request)
    pattern = re.compile(r'<div class="neirong">(.*?)</div>',re.S)
    lt = pattern.findall(content)
    text = lt[0]
    #将内容里面的图片标签全部清空
    pat = re.compile(r'<img .*?>')
    pat.sub('', text)
    return lt[0]

def parse_content(content):
    #写正则  想获取什么就对什么加小括号
    pattern = re.compile(r'<div class="item-t"><a href="/mrmy/lizhi/(\d+\.html)">(.*?)</a></div>', re.S)
    lt = pattern.findall(content)
    for href_title in lt:
        a_href = 'http://www.yikexun.cn/' + href_title[0]
        #获取标题
        title = href_title[-1]
        #向a_href发送请求，获取响应内容
        text = get_text(a_href)
        string = '<h1>%s</h1>%s' % (title, text)
        with open('lizhi.html', 'a', encoding = 'utf-8') as fp:
            fp.write(string.encode())
    
def main():
    url = 'http://www.yikexun.cn/mrmy/lizhi/list_54_'
    
    start_page = int(input('请输入开始页吗：'))
    end_page = int(input('请输入结束页吗：'))
    
    for page in range(start_page, end_page + 1):
        #根据url和page生成指定的request
        request = handle_request(url, page)
        #发送请求
        content = urllib.request.urlopen(request).read()
        #解析内容
        parse_content(content)

main()
```
